# Task 1

```python
import pandas as pd
```

```python
df1=pd.read_csv("nba.csv")
```

```python
df1.head()
```

# mean

```python
mean_age_by_team = df1.groupby('Team')['Age'].mean().reset_index()

print("Mean Age for Each Team:")
print(mean_age_by_team)
```

# median

```python
median_age_by_team = df1.groupby('Team')['Age'].median().reset_index()

print("Median Age for Each Team:")
print(median_age_by_team)
```

# max


```python
max_age_by_team = df1.groupby('Team')['Age'].max().reset_index()

print("Max Age for Each Team:")
print(max_age_by_team)
```

# min

```python
min_age_by_team = df1.groupby('Team')['Age'].min().reset_index()

print("Min Age for Each Team:")
print(min_age_by_team)
```

# standard deviation

```python
std_age_by_team = df1.groupby('Team')['Age'].std().reset_index()

print("std deviation Age for Each Team:")
print(std_age_by_team)
```
