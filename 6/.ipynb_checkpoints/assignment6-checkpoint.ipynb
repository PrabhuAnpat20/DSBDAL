```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Iris.csv")
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
sns.histplot(data=df, x='SepalLengthCm')
```

```python
sns.histplot(data=df, x='SepalWidthCm')
```

```python
sns.histplot(data=df, x='PetalLengthCm')
```

```python
sns.histplot(data=df, x='PetalWidthCm')
```

```python
sns.boxplot(data=df.drop(["Species"],axis=1))
```

```python
sns.boxplot(data=df, x="Species", y="SepalLengthCm")
```

```python
sns.boxplot(data=df, x="Species", y="SepalWidthCm")
```

```python
sns.boxplot(data=df, x="Species", y="PetalLengthCm")
```

```python
sns.boxplot(data=df, x="Species", y="PetalWidthCm")
```
